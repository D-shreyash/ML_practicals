```python
import pandas as pd 
```

```python
df = pd.read_csv("/home/shreyash/Desktop/ass4/dataset4.csv")
```

```python
df
```

```python
x= df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
x
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
```

```python
km=KMeans(n_clusters = 2)
```

```python
km.fit_predict(x)
```

```python
km.inertia_
```

```python
sse = []
for i in range(1,16):
    km = KMeans(n_clusters = i);
    km.fit_predict(x)
    sse.append(km.inertia_);
```

```python
sse
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("elbow method")
plt.xlabel("value of k")
plt.ylabel("sse")
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker = "." ,color ="red")
```

```python
km = KMeans(n_clusters = 5,random_state = 0)
```

```python
labels = km.fit_predict(x)
```

```python
labels
```

```python
plt.title("unclusterd data")
plt.xlabel("Anual Income")
plt.ylabel("spending score")
plt.scatter(df["Annual Income (k$)"],df["Spending Score (1-100)"])
```

```python
center = km.cluster_centers_
```

```python
center
```

```python
plt.title("clusterd data")
plt.xlabel("Annual Icome")
plt.ylabel("spending score")
plt.scatter(df["Annual Income (k$)"],df["Spending Score (1-100)"],c=labels)
plt.scatter(center[:,0],center[:,1],color ="black",s=100)
```

```python
kmeans = KMeans(n_clusters=5)
fit = kmeans.fit_predict(x)

# Plot data points for each cluster
plt.scatter(x.iloc[fit == 0, 0], x.iloc[fit == 0, 1], c='red', label='Cluster 0')
plt.scatter(x.iloc[fit == 1, 0], x.iloc[fit == 1, 1], c='green', label='Cluster 1')
plt.scatter(x.iloc[fit == 2, 0], x.iloc[fit == 2, 1], c='pink', label='Cluster 2')
plt.scatter(x.iloc[fit == 3, 0], x.iloc[fit == 3, 1], c='blue', label='Cluster 3')
plt.scatter(x.iloc[fit == 4, 0], x.iloc[fit == 4, 1], c='orange', label='Cluster 4')

plt.legend()
plt.show()
```
