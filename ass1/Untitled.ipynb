```python
list1=[]
for i in range(data.shape[0]):
    list1.append([str(data.values[i,j]) for j in range(data.shape[1]) if(str(data.values[i,j]) != 'nan')])
list1
```

```python
from  apyori import apriori
```

```python
rule = apriori(list1,min_support=0.005,min_confidence=0.2,min_lift=3,min_length=2)
result = __builtins__.list(rule)
```

```python
result
```

```python
list=[]
for item in result:
    pair = item[0]
    items = [x for x in pair]
    list.append([str(items[0]+" -> " + items[1]),str(item[1]),str(item[2][0][2]),str(item[2][0][3])])
```

```python
df= pd.DataFrame(list,columns=("rule","supp","conf","lift"))
```
